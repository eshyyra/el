```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = '/content/tested.csv'
```

```python
data = pd.read_csv(dataset)
```

```python
data
```

```python
# @title Sex

from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('Sex').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

dataset diatas berisikan tentang data penumpang titanic. dari dataset diatas yang diambil dari 418 sample data, diperoleh bahwa 150 penumpang titanic adalah perempuan yang ditandai dengan warna hijau dan sejumlah 253 penumpang lainnya adalah laki-laki yang ditandai dengan warna orange pada visualisasi tabel diatas.

```python
# @title Age

from matplotlib import pyplot as plt
data['Age'].plot(kind='hist', bins=20, title='Age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

tabel diatas adalah visualisasi dari umur penumpang titanic, data diatas menunjukkan bahwa penumpang dengan rentang umur 20 sampai 30 adalah penumpang terbanyak titanic, sedangkan untuk penumpang yang berumur lain memiliki frekuensi yang kurang dari 30, dan juga bisa diketahui dari tabel diatas bahwa tidak ada penumpang yang berumur sekitaran usia 70 tahun.


```python
# @title Survived

from matplotlib import pyplot as plt
data['Survived'].plot(kind='hist', bins=20, title='Survived')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

visualisasi data diatas menunjukkan data penumpang yang berhasil *survived* dan yang tidak berhasil untuk *survived*, didapatkan hasil bahwa sejumlah 150+ penumpang kapal titanic berhasil untuk *survived* yang ditunjukkan dengan angka 1.0 sedangkan terdapat 250+ penumpang dari dataset diatas yang tidak berhasil selamat yang ditunjukkan dengan angka 0.0.
